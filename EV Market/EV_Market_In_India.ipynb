#Principal Component Analysis (PCA)

Performing PCA to reduce the dimensionality of the feature space. This will help in visualize the data and potentially improve clustering results.

```python
from sklearn.decomposition import PCA

# Apply PCA
pca = PCA(n_components=9)  # You can choose the number of components
X_pca = pca.fit_transform(X)
```

```python
df2 = pd.DataFrame(X_pca, columns=['PC1', 'PC2','PC3','PC4','Pc5','PC6', 'PC7', 'PC8','PC9'])
df2
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage
linked = linkage(df2, 'complete')
plt.figure(figsize=(13, 9))
dendrogram(linked, orientation='top')
plt.show()
```

A dendrogram is a tree-like diagram that shows the arrangement of clusters and their relationships in a hierarchical clustering analysis. I have used the "complete" linkage method, which measures the dissimilarity between clusters as the maximum pairwise distance between their data points. The dendrogram will show how data points are organized into clusters based on this linkage criterion.

#K-Means Clustering using the Elbow Method

K-Means is an unsupervised machine learning algorithm that partitions a dataset into K clusters based on the similarity of data points. To find the optimal K for K-Means clustering, "Elbow Method used. Determining the optimal number of clusters (K) based on the reduced feature space after PCA

```python
from sklearn.cluster import KMeans

# Determine the optimal number of clusters (K) using the Elbow Method
inertia_values = []
k_range = range(2, 11)
for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=0)
    kmeans.fit(X_pca)
    inertia_values.append(kmeans.inertia_)

# Plot the Elbow Method
plt.plot(k_range, inertia_values, marker='o')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Inertia')
plt.title('Elbow Method for Optimal K (PCA)')
plt.show()
```

```python
# optimal K based on the Elbow Method is 4
optimal_k = 4
#K-Means clustering with the K=4
kmeans = KMeans(n_clusters=4, random_state=0)
clusters = kmeans.fit_predict(X_pca)
```

```python
print (kmeans.labels_) #Label assigned for each data point
print (kmeans.inertia_) #gives within-cluster sum of squares.
print(kmeans.n_iter_) #number of iterations that k-means algorithm runs to get a minimum within-cluster sum of squares
print(kmeans.cluster_centers_)
```

```python
# a scatter plot of the clusters
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=clusters, cmap='viridis')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.title('PCA Visualization of Clusters')
plt.colorbar(label='Cluster')
plt.show()
```

```python
# Group the data by cluster label
clustered_data = df1.copy()
clustered_data['Cluster'] = clusters

#cluster-wise statistics for key features
cluster_stats = clustered_data.groupby('Cluster').agg({
    'AccelSec': 'mean',
    'TopSpeed_KmH': 'mean',
    'Range_Km': 'mean',
    'Battery_Pack Kwh': 'mean',
    'Efficiency_WhKm' : 'mean',
    'FastCharge_KmH' : 'mean',
    'RapidCharge' : 'mean',
    'Seats' : 'mean',
    'PowerTrain' : 'mean',
    'INR' : 'mean'
})
print(cluster_stats)

# Visualize feature distributions by cluster : 'Range_Km' vs. 'Battery_Pack Kwh'
sns.scatterplot(data=clustered_data, x='Range_Km', y='Battery_Pack Kwh', hue='Cluster', palette='viridis')
plt.xlabel('Range (Km)')
plt.ylabel('Battery Pack (Kwh)')
plt.title('Feature Distribution by Cluster : Range_Km vs. Battery_Pack Kwh')
plt.show()

```
